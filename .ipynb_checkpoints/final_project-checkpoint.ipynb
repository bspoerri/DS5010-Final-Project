## About the data (wine)

These datasets are made up samples taken from red and white vinho verde wines. Vinho Verde is not a grape variety, but instead signifies a regional origin of the wine in the Minho Province of nothern Portugal. These wines are typically known for their "acidity" and "freshness," however, aromas and notes can vary with grape variety. Most Vinho Verde wines are "young," meaning they are released within a few months and consumed not long after being bottled.

## Field Descriptions

**fixed acidity**

   Acidity is fundamental to a wines flavor profile. Fixed acids (tartaric, malic, citric, and succinic) are those mostly originating from the grapes used in wine production. Warm climate grapes tend to have lower acidity. Low acidity can make a wine taste flat, but also may contribute to a softer taste when many prefer. Sometimes fixed acids are added or neutralized depending on desired characteristics.

   Units: g/L (tartaric acid) --> grams/Liter

**volatile acidity**

   Steam distillable acids present in wine (primarily acetic acid but also lactic, formic, butyric, and propionic acids) as a biproduct of mecrobial metabolism, not the grapes used to produce the wine, and are indicative of a "spoilage" product such as vinegar. They are minimized by reducing oxygen exposure and adding sulfur dioxide to preserve complexity and depth of wine flavor. Volatile acids can result in off-putting aromas. They may also disrupt the fermentation process.

   Units: g/L (acetic acid) --> grams/Liter

**citric acid**

   Citric acid is a "fixed acid" is often added to wines as a means of increasing acidity or complementing specific flavors. Cirtric acid can also improve a wines ferric stability (preventing excess iron which can result during formentation. The primary byproduct of citric acid is diacetyl which has a buttery odor in flavor. Too much citric acid can also increase the production of volatile acids which are generally not desired.

   Units: g/L --> grams/Liter

**residual sugar**

   Remaining naturally occurring sugar remaining in wine after the fermentation process. Wines are typically bucketed into "Bone Dry," "Dry," "Off-Dry," "Medium-Sweet," and "Sweet" categories (0-4 g/L, 5-11 g/L, 12-35 g/L, 35+ g/L respectively). During fermentation, yeast eats the sugar until the fermentation process is stopped. Thus, most sweet wines will have a lower alcohol content. 

   Units: g/L --> grams/Liter

**chlorides**

   Chloride concentration in wine is predominantly influenced by a wine's *Terrior* (properties of geographical origin and type of grape). Flavor is stringly impacted by chloride ions and in too high concentrations can result in an undesirable salty taste.

   Units: g/L (sodium chloride) --> grams/Liter
   
**free sulfur dioxide**

   Free (unbound) sulfur dioxide acts as an antioxidant, removing oxygen suspended in the wine. Reducing oxygen level slows down aging, reduces accumulation of volatile acids, and has an antimicrobial effect. Annecdotally, sulfites (sulfur dioxide) are attributed with causing headaches, however, these claims have not been corroborated by medical research.

   Units: mg/L --> milligrams/Liter

**total sulfur dioxide**

  Sum of free and molecularly bound sulfur dioxide. Note that only free sulfur dioxide has preservative properties. The maximum allowable concentration is 350 ppm (mg/L). The EU has lower maximums, generally 150 mg/L for red wine and 200 mg/L for white/rosé.

  Units: mg/L --> milligrams/Liter

```python
# quick check for any rows with a null value in any column
null_check = lambda df: df.isna().any(axis=1).sum()
print(f"Red Wine dataset has {null_check(red_wines)} of {len(red_wines)} rows with a null value.")
print(f"White Wine dataset has {null_check(white_wines)} of {len(white_wines)} rows with a null value.")
```

```python
# add type ("red" or "white") as a field in each independent dataset prior to joining
red_wines["type"] = "red"
white_wines["type"] = "white"
# concatenate data sets together-same table structures for each set
combined = pd.concat([red_wines, white_wines])
combined = combined.sort_values(by="quality", ascending=False)
```

```python
# calculate percentages of each type of wine
percentages = combined["type"].value_counts(normalize=True).round(3) * 100
plt.figure(figsize=(6,6))
plt.pie(percentages, 
        autopct = '%.1f%%',
        colors=['palegoldenrod', 'indianred'])
plt.title("Percentage of Samples by Wine Type")
plt.legend(labels=percentages.index, loc='lower right')
plt.savefig("problem_4_result.png")
```

## Combined Dataset Description

Combined, the two datasets contain 6,497 samples. Red wines make up about 1/4 of the total samples, with whites making up the remaining. These proportions are representative of the respective production volumes for each type in the region. 

The dataset has exceptional completeness with no missing physiochemical marker entries or ratings for any of the samples.

```python
# check the overall quality rating distribution by type of wine
rating_distribution = (combined.groupby(["type"])["quality"].value_counts()
                       .unstack(level=0)
                       .sort_index(ascending=True))
# plot the distribution of ratings by 
rating_distribution.plot(kind='bar', width=0.9,
                         xlabel="Quality", ylabel="Count", 
                         color=["indianred", "palegoldenrod"], 
                         title="General Distribution of Samples by Wine Type")
```

### 

```python
stat_summary = pd.concat([
    combined.groupby("type")["quality"].describe(),
    combined["quality"].describe().to_frame("combined").T
])
print(stat_summary)
```

```python
for wine_type, group in combined.groupby("type"):
    stat, p = stats.normaltest(group["quality"])
    print(f"{wine_type}: stat={stat:.4f}, p={p:.4f} → {'Normal' if p > 0.05 else 'Not normal'}")
```
